## 6. Create Voice Cloning Function

```python
import os
import time
import tempfile
from pathlib import Path

# Language mapping for XTTS-v2
LANGUAGES = {
    "English": "en",
    "Spanish": "es",
    "French": "fr",
    "German": "de",
    "Italian": "it",
    "Portuguese": "pt",
    "Polish": "pl",
    "Turkish": "tr",
    "Russian": "ru",
    "Dutch": "nl",
    "Czech": "cs",
    "Arabic": "ar",
    "Chinese": "zh-cn",
    "Japanese": "ja",
    "Hungarian": "hu",
    "Korean": "ko"
}

def clone_voice_modern(text, audio_file, language):
    """Modern voice cloning function with compatibility handling"""
    try:
        # Validate inputs
        if not text or not text.strip():
            return None, "❌ Please enter text to synthesize."

        if audio_file is None:
            return None, "❌ Please upload a reference audio file."

        # Check if audio file exists
        if not os.path.exists(audio_file):
            return None, "❌ Audio file not found. Please upload again."

        # Check if TTS is available
        if tts is None:
            return None, "❌ TTS model not loaded. Please check previous cell for errors."

        lang_code = LANGUAGES[language]

        # Create unique output filename
        timestamp = int(time.time())
        output_path = f"outputs/voice_clone_{timestamp}.wav"

        print(f"🎵 Generating {language} speech...")
        print(f"📝 Text: {text[:50]}...")
        print(f"🎤 Reference: {audio_file}")

        # Generate speech
        tts.tts_to_file(
            text=text.strip(),
            speaker_wav=audio_file,
            language=lang_code,
            file_path=output_path
        )

        # Check if output was created
        if os.path.exists(output_path):
            file_size = os.path.getsize(output_path)
            print(f"✅ Generated {file_size} bytes audio file")
            return output_path, f"✅ Speech generated successfully! ({file_size/1024:.1f} KB)"
        else:
            return None, "❌ Failed to generate speech file."

    except Exception as e:
        print(f"❌ Error in voice cloning: {str(e)}")
        return None, f"❌ Error: {str(e)}"

print("🔧 Voice cloning function ready!")
```

## 7. Build Modern Gradio Interface

```python
import gradio as gr

# Create interface with modern Gradio syntax
with gr.Blocks(
    title="XTTS2 Voice Cloning",
    theme=gr.themes.Soft()
) as demo:

    gr.HTML("""
    <div style="text-align: center; margin-bottom: 20px;">
        <h1>🎙️ XTTS2 Voice Cloning</h1>
        <p>Upload a 10-second audio sample and enter text to generate speech in the same voice</p>
        <p><strong>Python 3.12+ Compatible Version</strong></p>
    </div>
    """)

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 📤 Input Settings")

            # Audio input
            audio_input = gr.Audio(
                label="🎵 Reference Audio (10+ seconds recommended)",
                type="filepath"
            )

            # Language selection
            language_select = gr.Dropdown(
                choices=list(LANGUAGES.keys()),
                value="English",
                label="🌍 Output Language"
            )

            # Generate button
            generate_btn = gr.Button(
                "🎵 Generate Speech",
                variant="primary"
            )

            # Clear button
            clear_btn = gr.Button(
                "🗑️ Clear All",
                variant="secondary"
            )

        with gr.Column(scale=1):
            gr.Markdown("### 📝 Text Input")

            # Text input
            text_input = gr.Textbox(
                label="✍️ Text to Synthesize",
                placeholder="Enter the text you want to convert to speech...",
                lines=6
            )

            gr.Markdown("### 🎯 Output")

            # Output audio
            audio_output = gr.Audio(
                label="🔊 Generated Speech"
            )

            # Status
            status_output = gr.Textbox(
                label="📊 Status",
                interactive=False,
                lines=2
            )

    # Examples
    with gr.Row():
        with gr.Column():
            gr.Markdown("### 💡 Example Texts")
            gr.Examples(
                examples=[
                    ["Hello, this is a test of the XTTS voice cloning technology. I hope you find it impressive!", "English"],
                    ["Bonjour, je suis une voix générée par l'intelligence artificielle. C'est vraiment fascinant!", "French"],
                    ["Hola, esto es una prueba de clonación de voz. La tecnología es increíble!", "Spanish"]
                ],
                inputs=[text_input, language_select],
                label="Example texts (upload your own audio first)"
            )

    # Instructions
    with gr.Accordion("📖 Instructions & Tips", open=False):
        gr.Markdown("""
        ## 🚀 How to Use:
        1. **🎤 Upload Audio**: Upload a clear audio file (10+ seconds recommended)
        2. **🌍 Select Language**: Choose the language for synthesis
        3. **✍️ Enter Text**: Type or paste the text you want to convert
        4. **🎵 Generate**: Click the button to create speech
        5. **💾 Download**: Right-click the output audio to save

        ## 💡 Best Practices:
        - **High-quality audio**: Clear, no background noise
        - **Proper length**: 10-30 seconds works best
        - **Language matching**: Use same language as reference
        - **Text length**: Keep under 200 characters for speed
        - **Patience**: CPU inference takes ~10-15 seconds

        ## ⚠️ Troubleshooting:
        - **No audio generated**: Check reference audio quality and length
        - **Slow processing**: This is normal on CPU - be patient
        - **Memory issues**: Restart the runtime
        - **Poor quality**: Try better reference audio
        - **Model loading errors**: Check the model loading cell above
        """)

    # Event handlers
    generate_btn.click(
        fn=clone_voice_modern,
        inputs=[text_input, audio_input, language_select],
        outputs=[audio_output, status_output]
    )

    clear_btn.click(
        fn=lambda: ("", None, "English", None, "🗑️ Cleared all inputs"),
        outputs=[text_input, audio_input, language_select, audio_output, status_output]
    )

print("🎉 Modern interface created successfully!")
```
