## Step 1: Setup API Key

```python
# Install required libraries
!pip install requests

import requests
import json
from IPython.display import Audio, display
import os
from getpass import getpass

# Enter your API key securely
print("🔑 Enter your Mistral API key:")
print("(Get it from console.mistral.ai)")
api_key = getpass("API Key: ")

if not api_key.startswith("mistral-"):
    print("⚠️ API key should start with 'mistral-'")
else:
    print("✅ API key format looks correct!")

# Store for later cells
os.environ['MISTRAL_API_KEY'] = api_key
```

## Step 2: TTS Function

```python
def generate_speech(text, voice="casual_male", speed=1.0, language="en"):
    """Generate speech using Mistral's Voxtral TTS API"""
    
    # API endpoint
    url = "https://api.mistral.ai/v1/audio/speech"
    
    # Headers
    headers = {
        "Authorization": f"Bearer {os.environ['MISTRAL_API_KEY']}",
        "Content-Type": "application/json"
    }
    
    # Request payload
    payload = {
        "model": "mistral-tts-voxtral-2603",
        "input": text,
        "voice": voice,
        "response_format": "wav",
        "speed": speed
    }
    
    try:
        print(f"🎵 Generating: '{text}'")
        print(f"🎤 Voice: {voice}")
        
        # Make request
        response = requests.post(url, headers=headers, json=payload, timeout=60)
        
        if response.status_code == 200:
            print("✅ Success!")
            
            # Get audio info
            audio_size = len(response.content)
            print(f"📊 Audio size: {audio_size:,} bytes")
            
            # Save audio file
            filename = f"speech_{voice}_{len(text)}.wav"
            with open(filename, "wb") as f:
                f.write(response.content)
            print(f"💾 Saved: {filename}")
            
            # Play audio
            display(Audio(data=response.content, rate=22050))
            
            return response.content, filename
            
        else:
            print(f"❌ Error: {response.status_code}")
            try:
                error_data = response.json()
                print(f"Details: {error_data}")
            except:
                print(f"Response: {response.text[:200]}")
            return None, None
            
    except Exception as e:
        print(f"❌ Request failed: {e}")
        return None, None

# Available voices
AVAILABLE_VOICES = [
    "casual_male", "casual_female",
    "professional_male", "professional_female",
    "narrator_male", "narrator_female",
    "friendly_male", "friendly_female",
    "calm_male", "calm_female"
]

print(f"🎤 Available voices: {', '.join(AVAILABLE_VOICES)}")
print("✅ TTS function ready!")
```

## Step 3: Test Examples

```python
# Test different voices and languages
test_cases = [
    {
        "text": "Hello! This is Mistral's Voxtral TTS using the API.",
        "voice": "casual_male",
        "lang": "English"
    },
    {
        "text": "Bonjour! Je suis Voxtral TTS de Mistral.",
        "voice": "casual_female",
        "lang": "French"
    },
    {
        "text": "¡Hola! Soy Voxtral TTS de Mistral.",
        "voice": "professional_male",
        "lang": "Spanish"
    },
    {
        "text": "Hallo! Ich bin Voxtral TTS von Mistral.",
        "voice": "friendly_female",
        "lang": "German"
    },
    {
        "text": "Ciao! Sono Voxtral TTS di Mistral.",
        "voice": "calm_male",
        "lang": "Italian"
    }
]

print("🎤 Testing Voxtral TTS API...")
print("⏳ Each request takes 5-15 seconds\n")

for i, test in enumerate(test_cases):
    print(f"--- Test {i+1}: {test['lang']} ---")
    
    audio_data, filename = generate_speech(
        text=test["text"],
        voice=test["voice"]
    )
    
    if audio_data:
        print(f"✅ {test['lang']} test successful!\n")
    else:
        print(f"❌ {test['lang']} test failed!\n")
    
    import time
    time.sleep(2)  # Brief pause between tests
```

## Step 4: Generate Custom Text

```python
# Generate your own text
print("🎤 Generate Custom Speech")
print("Enter your text and choose a voice:\n")

# Get user input (in Colab, you'll modify these variables)
custom_text = "This is a custom text generated with Mistral's Voxtral TTS API. It works perfectly without any GPU requirements!"
selected_voice = "casual_female"  # Change this to any voice from the list
custom_speed = 1.0  # 0.5 to 2.0, where 1.0 is normal

print(f"📝 Text: '{custom_text}'")
print(f"🎤 Voice: {selected_voice}")
print(f"⚡ Speed: {custom_speed}")
print("\n🎵 Generating...")

# Generate speech
audio_data, filename = generate_speech(
    text=custom_text,
    voice=selected_voice,
    speed=custom_speed
)

if audio_data:
    print(f"\n🎉 Custom speech generated successfully!")
    print(f"💾 File: {filename}")
    print("\n💡 You can download the audio file from the Colab file browser!")
else:
    print("\n❌ Custom speech generation failed!")
```
